Számítsuk ki, hogy az egyes vásárlások között hány nap telt el.

```sql
select *,
    lag(OrderDate, 1, OrderDate) over(order by OrderDate) as [Lag], --előző sor értékeit adja vissza    
    DATEDIFF(day, orderDate, lag(OrderDate,1 , '2000-01-01') over(order by OrderDate)) as DaysBetweenOrders,
    --cast((OrderDate - lag(OrderDate,1 , '2000-01-01') over(order by OrderDate)) as int) as LagWithDefault, --előző sor értékeit adja vissza
    lead(OrderDate) over(order by OrderDate) as [Lead], --következó sor értékeit adja vissza    
    FIRST_VALUE(SubTotal) over(Partition by CustomerID order by (select null)) as [First_Value] --kötelező order by esetén jó ez, olyan mintha nem adtad volna meg
from Orders
where CustomerID = 29734
order by OrderDate
```

```sql
SELECT *, DATEDIFF(day, o.orderDate, o.NextOrderDate) as Diff
FROM (
    SELECT
        OrderID,
        CustomerID,
        OrderDate,
        LEAD(OrderDate) over (partition by CustomerID order by OrderDate) as NextOrderDate
    FROM Orders
) o
ORDER BY CustomerID, OrderDate, OrderID
```

A lag paraléterei : mezőnév, hány sorral menjen vissza, default érték hiányzó / null esetén

Ha a where-t kiveszed, akkor célszerű a partition by -nál a customerID-t megadni, mert nem lesz helyes a futási eredmény

mer a sorok nem rendezetten vannak a táblában.

```sql
SELECT
    OrderID, OrderDate, SubTotal,
    lead(SubTotal)          over(partition by CustomerID order by OrderID) as [LEAD],
    --következő sor subtotalt adja vissza
    lag(SubTotal)           over(partition by CustomerID order by OrderID) as [LAG],
    --előző sor subtotalt adja vissza
    First_Value(SubTotal)   over(partition by CustomerID order by OrderID) as [FIRST_VALUE1],
    --rendezettség alapján az elsőt
    last_value(SubTotal)    over(partition by CustomerID order by OrderID desc) as [FIRST_VALUE2],
     --rendezettség alapján az utolsót
    --ha csökkenőbe rendezem ugyanaz mint a first value
    First_Value(SubTotal)   over(partition by CustomerID order by OrderID rows between unbounded preceding and unbounded following) as [FIRST_VALUE3],
    last_value(SubTotal)    over(partition by CustomerID order by OrderID) as [LAST_VALUE1],
    --azért nem jó, mert az ablak változik a lekérdezés alatt és a pillanatnyi utólsó értéket adja vissza
    First_Value(SubTotal)   over(partition by CustomerID order by OrderID desc) as [LAST_VALUE2],
    last_value(SubTotal)    over(partition by CustomerID order by OrderID rows between unbounded preceding and unbounded following) as [LAST_VALUE3]
FROM Orders
WHERE CustomerID = 11000
ORDER BY OrderID
```
